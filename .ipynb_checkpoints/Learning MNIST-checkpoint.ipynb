## MNIST

### Extract mnist dataset from sklearn lib

```python
from sklearn.datasets import fetch_openml
```

```python
mnist = fetch_openml('mnist_784', version=1)
mnist.keys()
```

### Extract array data and associated label

```python
X, y = mnist["data"].to_numpy(), mnist["target"].to_numpy()
```

```python
print(f"Num images: {X.shape[0]}, Num features (pixels): {X.shape[1]}")
```

### Display a sample of the dataset

```python
import matplotlib.pyplot as plt
```

```python
# display sample of digits
fig, axis = plt.subplots(10, 10, figsize=(5, 5))
labels = ""
for i, ax in enumerate(axis.flat):
    example_digit = X[i].reshape(28, 28)
    labels = labels + f"{y[i]}, "
    ax.imshow(example_digit, cmap='binary');
    ax.axis("off")
print(labels)
```

### Create training & test subsets
training = first 60,0000; test = last 10,000 images

note: the training set is already shuffled

```python
X_train, X_test, y_train, y_test = X[:60_000], X[60_000:], y[:60_000], y[60_000:]
```

### Support Vector Machine (SVM) classifier - Multiclass classification

```python
from sklearn.svm import SVC
from sklearn.model_selection import cross_val_score
```

```python
svm_clf = SVC(random_state=42)
svm_clf.fit(X_train[:2000], y_train[:2000])
```

```python
digit_to_predict = 499

digit_array = X[digit_to_predict].reshape(28, 28)
digit_label = y[digit_to_predict]
digit_prediction = svm_clf.predict([X[digit_to_predict]])

print(f"{digit_prediction[0]==digit_label}, Prediction: {digit_prediction[0]}, Actual: {digit_label}")

import numpy as np

some_digit_scores = svm_clf.decision_function([X[digit_to_predict]])
print(f"\nIndex: {np.argmax(some_digit_scores)}, Max score: {np.max(some_digit_scores).round(2)}")
print("Classes: ", svm_clf.classes_)
dec_func = svm_clf.decision_function([X[digit_to_predict]])
print("SVM Score per class for selected value: \n", dec_func)

plt.figure(figsize=(2,2))
plt.imshow(digit_array, cmap='binary')
plt.axis("off")
plt.show()
```

Score is above 89%. But we should do a scaling transformation to scale the data to get a better result. This will scale the matrix values from 0-255 (grey scale) to 0-1 to reduce bias in the model.

### Training an SGDClassifier (or RandomForestClassifier)

Create smaller subset for reduced training time

```python
import warnings
from sklearn.exceptions import ConvergenceWarning
warnings.simplefilter("ignore", category=ConvergenceWarning)
```

```python
X_train_sub, y_train_sub = X_train[:2000], y_train[:2000]
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=42)
sgd_clf.fit(X_train_sub, y_train_sub)
```

Testing the model with a random value and extracting the class score

```python
i = 485
prediction = sgd_clf.predict([X[i]])
label = y[i]
print(f"{prediction==label}, Prediction: {prediction}, Actual: {label}")
class_score = sgd_clf.decision_function([X[i]]).round()
print(class_score)
```

Evaluating the model

```python
# cross validation to evaluate the model
cross_val_score(sgd_clf, X_train_sub, y_train_sub, cv=3, scoring="accuracy")
```

The model gets at least 85.8% on all it's test folds. Using a random classifier would get 10% accuracy so this not bad. 

Using scaling to get a quick improvement in accuracy.

**Remember:** StandardScaler will scale the trainng data values i.e., from 0-255 (grey scale values) to 0-1. This reduces bias in the model towards larger values.

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train_sub.astype("float64"))
```

```python
cross_val_score(sgd_clf, X_train_scaled, y_train_sub, cv=3, scoring="accuracy")
```

The above normalised matrix shows that only 72% of the images of 5s were classified correctly. This model most often confused 5s with 3s and 7s (7% each).

To make errors stand out more we can put zero weighting on the correct predictions

```python
sample_weight = (y_train_pred != y_train_sub)
ConfusionMatrixDisplay.from_predictions(y_train_sub, y_train_pred, sample_weight=sample_weight, normalize="true", values_format=".0%")
plt.show()
```

The above is now normalised by rows (or labels) so we can now say 62% of the errors the model made on images of 0s were misclassifications as 5s.But in reality on 5% of images of 0s were misclassified as 5s

```python
ConfusionMatrixDisplay.from_predictions(y_train_sub, y_train_pred, sample_weight=sample_weight, normalize="pred", values_format=".0%")
plt.show()
```

The above confusion matrix normalises by the columns (or predictions). Now we can say 46% of misclassified 7s were actually 9s
